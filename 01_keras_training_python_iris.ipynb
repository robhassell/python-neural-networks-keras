```python
from keras.datasets import boston_housing
```

```python
data = boston_housing.load_data()
```

```python
x_train, y_train = data[0][0], data[0][1]
x_test, y_test = data[1][0], data[1][1]
```

```python
from keras.models import Sequential
from keras.layers import Activation, Dense
```

```python
model = Sequential()
model.add(Dense(units=13, input_dim=13, kernel_initializer='normal', activation='relu'))
```

```python
model.add(Dense(units=6, kernel_initializer='normal', activation='relu'))
```

```python
model.add(Dense(units=4, kernel_initializer='normal', activation='relu'))
```

```python
model.add(Dense(units=1, kernel_initializer='normal'))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_absolute_percentage_error'])
```

```python
model.summary()
```

```python
from keras.utils import plot_model

#plot_model(model)
```

```python
x_val = x_train[300:,]
y_val = y_train[300:,]
```

```python
history = model.fit(x=x_train, y=y_train, batch_size=32, epochs=100, validation_data=(x_val, y_val), verbose=0)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['mean_absolute_percentage_error'])
```
